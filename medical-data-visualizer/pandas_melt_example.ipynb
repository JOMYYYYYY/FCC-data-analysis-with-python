```python
import pandas as pd

data = pd.DataFrame({
    'ID': [101, 102, 103, 104],
    'Name': ['Alice', 'Bob', 'Charlie', 'David'],
    'Math_Midterm': [85, 78, 92, 88],
    'Math_Final': [90, 85, 96, 92],
    'Science_Midterm': [88, 89, 94, 90],
    'Science_Final': [92, 91, 98, 94]
})
data
```

```python
melted_data = pd.melt(data, id_vars=['ID', 'Name'], var_name='Exam_Type', value_name='Score')
melted_data
```
