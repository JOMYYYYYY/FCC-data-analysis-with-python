# Sea Level Change Analysis

In this analysis, you will work with a dataset that contains global average sea level change data since 1880. Your goal is to predict the sea level change through the year 2050.

## Tasks

1. Import Data
   - Use Pandas to import the data from the `epa-sea-level.csv` file.

2. Create a Scatter Plot
   - Use Matplotlib to create a scatter plot with the Year column as the x-axis and the CSIRO Adjusted Sea Level column as the y-axis.

3. Calculate Line of Best Fit
   - Use the `linregress` function from `scipy.stats` to calculate the slope and y-intercept of the line of best fit.

4. Plot the Line of Best Fit
   - Overlay the line of best fit on top of the scatter plot. Extend the line to go through the year 2050 to predict the sea level rise in 2050.

5. Calculate and Plot Recent Trends
   - Calculate a new line of best fit using data from the year 2000 through the most recent year in the dataset.
   - Plot this line to also go through the year 2050, predicting the sea level rise if the rate continues as it has since the year 2000.

## Plot Details

- X-label: Year
- Y-label: Sea Level (inches)
- Title: Rise in Sea Level

## Saving and Returning Results

The boilerplate code includes commands to save and return the image of the generated plot.


```python
import pandas as pd
import matplotlib.pyplot as plt
from scipy.stats import linregress

def draw_plot():
    # Read data from file
    data = pd.read_csv('epa-sea-level.csv')

    # Create scatter plot
    plt.scatter(data['Year'], data['CSIRO Adjusted Sea Level'], label='Sea Level Data')

    # Create first line of best fit
    slope, intercept, r_value, p_value, std_err = linregress(data['Year'], data['CSIRO Adjusted Sea Level'])
    line1 = [slope * x + intercept for x in range(1880, 2051)]
    plt.plot(range(1880, 2051), line1, label='Line of Best Fit (1880-2050)', color='orange')

    # Create second line of best fit (from 2000 to the most recent year)
    recent_data = data[data['Year'] >= 2000]
    slope, intercept, r_value, p_value, std_err = linregress(recent_data['Year'], recent_data['CSIRO Adjusted Sea Level'])
    line2 = [slope * x + intercept for x in range(2000, 2051)]
    plt.plot(range(2000, 2051), line2, label='Line of Best Fit (2000-2050)', color='green')

    # Add labels and title
    plt.xlabel('Year')
    plt.ylabel('Sea Level (inches)')
    plt.title('Rise in Sea Level')
    plt.legend()

    # Save plot and return data for testing (DO NOT MODIFY)
    plt.savefig('sea_level_plot.png')
    return plt.gca()
```

```python
fig = draw_plot()
```
